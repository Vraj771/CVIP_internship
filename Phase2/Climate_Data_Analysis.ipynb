```python
tmipd = pd.to_datetime(ddf["DATE"])
tmipd
```

```python
tm = pd.to_datetime(ddf["DATE"])
ddf.groupby(tm)['DailyAverageDryBulbTemperature'].mean().plot(xlabel='DATE')
```

```python
ddf['Season'] = (tmipd.dt.month % 12 + 3) // 3
seasonl_temerature = ddf.groupby('Season')['DailyAverageDryBulbTemperature'].mean()
```

```python
plt.figure(figsize=(12,6))
plt.plot(seasonl_temerature.index, seasonl_temerature.values,marker='o')
plt.title("Seasonl Average Temperature Trends")
plt.xlabel('Season')
plt.ylabel('Average Temperature (*F)')
plt.xticks([1,2,3,4],['Winter','Spring','Summer','Fall'])
plt.show()
```
